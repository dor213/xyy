```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/29943/Desktop/模式识别/实验数据集/Iris数据集/iris.csv")
data.head()
```

```python
label = data['Species'].unique()
label
```

```python
#选取每个类别的前30个样本作为训练集，后20个样本作为测试集
train_data = pd.DataFrame()
test_data = pd.DataFrame()
for i in label:
    train_data = train_data.append(data[data['Species']==i][:30])
    test_data = test_data.append(data[data['Species']==i][30:])

train_data = train_data.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)

train_data.head()
test_data.head()
```

```python
#使用pla算法分类
def pla(data, label , epoch_num):
    w = np.zeros(len(data[0]))
    epoch = 0
    while epoch < epoch_num:
        for i in range(len(data)):
            if np.dot(w, data[i])*label[i] <= 0:
                w = w + data[i]*label[i]
        epoch += 1
    return w

```

```python
#以pla为基础分类算法，对iris数据集OVO分类


#对训练样本进行重新分组
train_data_ovo = pd.DataFrame()
train_data_ovo_1 = train_data[train_data['Species']=='setosa']
train_data_ovo_2 = train_data[train_data['Species']=='versicolor']
train_data_ovo_3 = train_data[train_data['Species']=='virginica']

train_data_ovo_1['Species'] = 1
train_data_ovo_2['Species'] = 2
train_data_ovo_3['Species'] = 3

#对类别1和类别2进行分类
train_data_ovo_12 = train_data_ovo_1.append(train_data_ovo_2)
train_data_ovo_12 = train_data_ovo_12.reset_index(drop=True)

#对类别1和类别3进行分类
train_data_ovo_13 = train_data_ovo_1.append(train_data_ovo_3)
train_data_ovo_13 = train_data_ovo_13.reset_index(drop=True)

#对类别2和类别3进行分类
train_data_ovo_23 = train_data_ovo_2.append(train_data_ovo_3)
train_data_ovo_23 = train_data_ovo_23.reset_index(drop=True)

#train_data_ovo_12中类别1为1，类别2为-1
train_data_ovo_12.loc[train_data_ovo_12['Species']==1, 'Species'] = 1
train_data_ovo_12.loc[train_data_ovo_12['Species']==2, 'Species'] = -1

#train_data_ovo_13中类别1为1，类别3为-1
train_data_ovo_13.loc[train_data_ovo_13['Species']==1, 'Species'] = 1
train_data_ovo_13.loc[train_data_ovo_13['Species']==3, 'Species'] = -1

#train_data_ovo_23中类别2为1，类别3为-1
train_data_ovo_23.loc[train_data_ovo_23['Species']==2, 'Species'] = 1
train_data_ovo_23.loc[train_data_ovo_23['Species']==3, 'Species'] = -1

train_data_ovo_12
```

```python
#使用softmax算法对iris数据集OVO分类
#获取训练样本的特征和标签，并将其转换为numpy数组
train_data_feature = train_data.iloc[:, :-1].values
train_data_label = train_data.iloc[:, -1].values

#获取测试样本的特征和标签，并将其转换为numpy数组
test_data_feature = test_data.iloc[:, :-1].values
test_data_label = test_data.iloc[:, -1].values

#训练样本的标签编号替换为1，2，3
train_data_label[train_data_label=='setosa'] = 1
train_data_label[train_data_label=='versicolor'] = 2
train_data_label[train_data_label=='virginica'] = 3

#测试样本的标签编号替换为1，2，3
test_data_label[test_data_label=='setosa'] = 1
test_data_label[test_data_label=='versicolor'] = 2
test_data_label[test_data_label=='virginica'] = 3


#定义交叉熵损失函数
def cross_entropy_loss(w, x, y):
    loss = 0
    for i in range(len(x)):
        loss += np.log(1+np.exp(-y[i]*np.dot(w, x[i])))
    return loss

#搭建softmax分类器
def softmax(x):
    x = np.exp(x)
    return x/np.sum(x)


#使用softmax算法对iris数据集分类
#初始化参数w,w为3*4的矩阵
w = np.zeros((3, 5))
loss = []
epoch = 0
while epoch < 1000:
    for i in range(len(train_data_feature)):
        if train_data_label[i] == 1:
            y = np.array([1, -1, -1])
            # print(softmax(np.dot(w[0], train_data_feature[i])))
            w[0] = w[0] - 0.01*softmax(np.dot(w[0], train_data_feature[i]))*train_data_feature[i]


        elif train_data_label[i] == 2:
            y = np.array([-1, 1, -1])
            w[1] = w[1] - 0.01*softmax(np.dot(w[1], train_data_feature[i]))*train_data_feature[i]

        else:
            y = np.array([-1, -1, 1])
            w[2] = w[2] - 0.01*softmax(np.dot(w[2], train_data_feature[i]))*train_data_feature[i]
        
    epoch += 1
print('softmax分类器的参数为：', w)

#对测试样本进行分类
correct_num = 0
for i in range(len(test_data_feature)):
    if test_data_label[i] == 1:
        y = np.array([1, -1, -1])
    elif test_data_label[i] == 2:
        y = np.array([-1, 1, -1])
    else:
        y = np.array([-1, -1, 1])
    if np.argmax(softmax(np.dot(w, test_data_feature[i]))) == np.argmax(y):
        correct_num += 1
correct_rate = correct_num/len(test_data_feature)
print('softmax分类器的正确率为：', correct_rate)

 

```
